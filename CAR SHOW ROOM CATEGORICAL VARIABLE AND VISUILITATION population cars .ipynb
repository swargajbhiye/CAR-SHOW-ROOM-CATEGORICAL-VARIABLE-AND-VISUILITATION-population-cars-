```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# define data for cars brands and their frequencies
```

```python
car_data = {
    'Brand': ['Audi','BMW','Mercedes'],
    'Frequency': [124, 98, 113]
}
```

```python
#create a data frame
```

```python
df = pd.DataFrame(car_data)
total_frequency = df['Frequency'].sum()
```

```python
#Display table of freequencies
```

```python
print("Table of Car Frequencies:")
print(df)
print("\nTotal Frequency:", total_frequency)
```

```python
# Plotting the bar chart
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df['Brand'], df['Frequency'] , color=['red','green','blue'])
plt.title(" CAR Brand Frequency at German Car Shop")
plt.xlabel("Car Brand")
plt.ylabel("Frequency")
plt.show()
```
